```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen as uReq
import logging
import os
```

```python
save_dir = "images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.56"}
```

```python
query = "Virat Kohli"
response = requests.get(f"https://images.search.yahoo.com/search/images?p={query}&fr=mcafee")
```

```python
Soup = BeautifulSoup(response.content, 'html.parser')
print(Soup)
```

```python
images_tags = Soup.find_all("img")
```

```python
del images_tags[0]
```

```python
img_data_mongo = []
for i in images_tags:
    image_url = i.get("src") or i.get("data-src") or i.get("data-original")
    image_data = requests.get(image_url).content
    mydict = {"index":image_url , "image" : image_data}
    img_data_mongo.append(mydict)

    with open(os.path.join(save_dir, f"{query}_{images_tags.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
```
